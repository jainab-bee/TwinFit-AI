```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df["label"].value_counts().plot(kind="bar")

plt.title("Label Distribution")
plt.xlabel("Label")
plt.ylabel("Count")

plt.show()
```

```python
df["resume_length"] = df["resume_text"].astype(str).apply(len)

print(df["resume_length"].describe())
```

```python
plt.hist(df["resume_length"], bins=30)

plt.title("Resume Length Distribution")
plt.xlabel("Characters")
plt.ylabel("Count")

plt.show()
```

```python
df["jd_length"] = df["job_description_text"].astype(str).apply(len)

print(df["jd_length"].describe())
```
